```python
from pdfplumber import pdfplumber
from operator import itemgetter

IN_PATH = 'nics-background-checks-single-page.pdf'

the_pdf = pdfplumber.open(IN_PATH)
pg = the_pdf.pages[0]

```

```python
crop_month = pg.within_bbox((0, 35, pg.width, 65))x
crop_month.to_image(resolution=150)
```

```python
crop_table = pg.crop((0, 80, pg.width, 485))
crop_table.to_image(resolution=150)
```

```python
_cfg = {
        "intersection_tolerance": 5,

        "horizontal_strategy": "text",
        "intersection_tolerance": 5

}

crop_table.to_image().debug_tablefinder(_cfg)
```

```python
_cfg = {
        "horizontal_strategy": "text",
        "explicit_vertical_lines": [
            min(map(itemgetter("x0"), crop_table.chars))
        ],
        "intersection_tolerance": 5
}

crop_table.to_image().debug_tablefinder(_cfg)
```
