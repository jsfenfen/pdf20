The point of this exercise is to get a more useful result than what you can already get through a program like ABBYY Finereader ([example outpupt](abbyyfr-gsk-grants-2017.xlsx)


```python
from pdfplumber import pdfplumber

IN_PATH = 'gsk-grants-2017.pdf'

the_pdf = pdfplumber.open(IN_PATH)
```

```python
# one page
pg = the_pdf.pages[0]

px = pg.crop((0, 95, pg.width, pg.height))
px.to_image()
```
